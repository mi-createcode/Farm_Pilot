```python
import pandas as pd
from collections import defaultdict

# ì—‘ì…€ ì½ê¸° (ë‘ ë²ˆì§¸ ì¤„ë¶€í„° ë°ì´í„° ì‹œì‘ì´ë©´ header=1)
df = pd.read_excel("jeonbuk.xlsx", header=1)
df.fillna("", inplace=True)

# ì—´ ì´ë¦„ ì§€ì •
df.columns = ["ë„", "ì‹œ", "êµ¬", "ë™"]

# 4ë‹¨ê³„ ë”•ì…”ë„ˆë¦¬ ë§Œë“¤ê¸°: ë„ â†’ ì‹œ â†’ êµ¬ â†’ ë™ ë¦¬ìŠ¤íŠ¸
region_data = defaultdict(lambda: defaultdict(lambda: defaultdict(list)))

for _, row in df.iterrows():
    do, si, gu, dong = row["ë„"], row["ì‹œ"], row["êµ¬"], row["ë™"]
    if dong:
        region_data[do][si][gu].append(dong)
    else:
        # ë™ì´ ì—†ë”ë¼ë„ êµ¬ì¡°ë§Œ ì¡ê¸°
        _ = region_data[do][si][gu]

# ì¼ë°˜ dictë¡œ ë³€í™˜
region_data = {
    do: {
        si: dict(gu_dict)
        for si, gu_dict in si_dict.items()
    }
    for do, si_dict in region_data.items()
}
import pprint
pprint.pprint(region_data)
```
